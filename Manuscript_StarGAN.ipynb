```python
import os
import matplotlib.pyplot as plt
from PIL import Image

# Define the base directory
base_dir = "/home/mikylab/manuscript/stargan/experiments/stargan_identity.bel01/results"

# Define the number of predictions
num_predictions = 3
labels = [2121.6484, 5149.591, 4332.4634]

# Mapping numbers to words
mapping = {0: 'Blast', 1: 'Erythroblast', 2: 'Lymphoctye'}

# Create a figure with subplots
fig, axes = plt.subplots(nrows=1, ncols=num_predictions, figsize=(12, 8))

# Iterate over predictions
for pred in range(num_predictions):
    # Construct the image path
    image_path = os.path.join(base_dir, f"class0_img0_pred{pred}.jpg")

    # Load image
    img = Image.open(image_path)

    # Set title and label
    title = f"Class: {mapping[0]}\nPrediction: {mapping[pred]}"
    label = f"Translation Distance: {labels[pred]:,.2f}"  # Format label with comma for thousands and two decimal places

    # Plot the image along with its title and label
    ax = axes[pred]
    ax.imshow(img)
    ax.set_title(f"{title}\n{label}")
    ax.axis('off')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn import linear_model
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import svm


def normalizeData(dataMatrix):
    distance_mean = np.mean(dataMatrix)
    distance_sd = np.std(dataMatrix)
    mean_zero = (dataMatrix-distance_mean)/distance_sd
    return mean_zero
def accuracy(truth, predict):
    return (np.sum(truth == predict))/truth.shape[0]

trainDistances = pd.read_csv('/home/mikylab/manuscript/stargan/experiments/stargan_identity.class6/train_distances.csv', header = None)
testDistances = pd.read_csv('/home/mikylab/manuscript/stargan/experiments/stargan_identity.class6/test_distances.csv', header = None)
```

```python
'''
Standardize the data so that it can be read in and used for the models.
'''
trainArray = np.asarray(trainDistances)
trainMatrix = trainArray[:, 1:]

X_train = normalizeData(trainMatrix)
Y_train = trainArray[:, 0]

testArray = np.asarray(testDistances)
testMatrix = testArray[:, 1:7]


X_test = normalizeData(testMatrix)
Y_test = testArray[:, 0]
```

```python
''' Smallest Translation Distance'''
min_x_train = X_train.argmin(axis = 1)
min_x_test = X_test.argmin(axis = 1)

tl_train_acc = accuracy(Y_train, min_x_train)
tl_acc = accuracy(Y_test, min_x_test)

tl_acc
```

```python
testDistances['predicted_class'] = min_x_test
```
